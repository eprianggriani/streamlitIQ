1. Tentukan Library yang akan digunakan

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

3. Standarisasi Data

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
standarized_data = scaler.transform(X) 
```

```python
print(standarized_data)
```

```python
X = standarized_data
Y=iq_rn[['Nilai IQ', 'Outcome']]
```

```python
print(X)
print(Y)
```

4. Memisahkan Data Training dan Data Testing

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

5. Membuat data latih menggunakan algoritma Random Forest

```python
#Data untuk training dan test regresi (mendadapatkan nilai IQ)
Y_train_IQ = Y_train['Nilai IQ']
Y_test_IQ = Y_test['Nilai IQ']

print(Y_train_IQ)
```

```python
#Data untuk training dan test Klasifikasi
Y_train_outcome = Y_train['Outcome']
Y_test_outcome = Y_test['Outcome']

print(Y_train_outcome)
```

```python
regresi_nilai = LinearRegression()
regresi_nilai.fit(X_train,Y_train_IQ)
```

```python
classifier = RandomForestClassifier(n_estimators=100, random_state=42)
classifier.fit(X_train, Y_train_outcome)
```

6. Membuat model evaluasi untuk mengukur tingkat akurasi

```python
Y_trn_IQ = regresi_nilai.predict(X_train)
mse_trn = mean_squared_error(Y_train_IQ, Y_trn_IQ)
print(f"MSE data Training Nilai IQ: {mse_trn}")
```

```python
Y_trn_outcome = classifier.predict(X_train)
accuracy = accuracy_score(Y_train_outcome, Y_trn_outcome)
print(f"Akurasi data training Outcome adalah {accuracy}")
```

```python
Y_pred_IQ = regresi_nilai.predict(X_test)
mse = mean_squared_error(Y_test_IQ, Y_pred_IQ)
print(f"MSE data testing Nilai IQ adalah {mse}")
```

```python
Y_pred_outcome = classifier.predict(X_test)
accuracy = accuracy_score(Y_test_outcome, Y_pred_outcome)
print(f"Akurasi data testing Outcome adalah {accuracy}")
```

7. Membuat model prediksi

```python
input_data = (90)

input_data_as_numpy_array = np.array(input_data)

input_data_reshape = input_data_as_numpy_array.reshape(1,-1)

std_data = scaler.transform(input_data_reshape)
print(std_data)

pred= regresi_nilai.predict(std_data)
prediksi_iq = round(pred[0])
print('Nilai IQ:', prediksi_iq)

prediction = classifier.predict(std_data)
print('Outcome;', prediction)

if (prediction[0] == 1):
    print('Dibawah rata-rata')
elif (prediction [0] == 2):
    print ('Rata-rata')
else :
    print('diatas rata-rata') 
```
